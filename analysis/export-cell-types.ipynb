# Purpose
This document extracts the annotated cell types from [the Human Cell Landscape (Han et al., *Nature*, 2020)](https://doi.org/10.1038/s41586-020-2157-4) dataset.
## Load modules

```python
import pandas as pd
import scanpy as sc
from matplotlib.pyplot import rc_context
```

## Load data

```python
## data available at: https://db.cngb.org/HCL/data/HCL_final_USE.h5ad.gz
hcl = sc.read_h5ad("../data/scanpy/HCL_final_USE.h5ad", backed='r')
hcl
```

## Identify cell types

```python
with rc_context({'figure.figsize': (12, 6)}):
    sc.pl.violin(hcl, ['FXR1'], groupby="louvain", rotation=90, log=True)
```
